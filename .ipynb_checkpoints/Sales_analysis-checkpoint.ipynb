### Read in updated dataframe

```python
all_data = pd.read_csv("all_months_data.csv")
all_data.head()
```

```python
all_data.shape
```

```python
all_data['Product'].dtype
```

## Clean Up the Data !

#### Drop Rows of NaN

```python
#Checking all the NaN values in the dataset
all_data[all_data.isna().any(axis=1)]
```

```python
# Dropping all the NaN values from the Dataset using  .dropna() 
all_data = all_data.dropna()
all_data
```

```python
#Checking if there is any NaN values in the Dataset
all_data.isna().sum()
```

```python
#Removing all the rows that does not have any date on the order date column
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

### Convert all the columns to the correct types :

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

### Augment data with additional columns

### Task 2: Add month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### Task 3 : Add a Sales Column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each'] 

all_data.head()
```

### Task 4 : Adding a City Column 

```python
#Using .apply() function

def get_city(address):                       # Using function to split the city name 
    return address.split(',')[1]
def get_state(address):                          # Using function to split the state name
    return address.split(',')[2][1:3]

city_list = all_data['Purchase Address'].apply(lambda x: f'{get_city(x)} ({get_state(x)})')
# city_list
```

```python
all_data['City Name'] = city_list 
```

### Data Exploration!

#### Question 1: What was the best month for sales? How much was earned that month? 

```python
sales_data = all_data.groupby('Month').sum()['Sales']
sales_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

months = range(1,13)
plt.bar(months, sales_data)
plt.xticks(months)
plt.ylabel('Sales in USD($)')
plt.xlabel('Month Number')
plt.show()
```

### Question 2: What US city had the highest number of sales ?

```python
all_data.head()
```

```python
# Grouping by the city names in order to see the highest sales in city

highest_sales = all_data.groupby('City Name').sum('Sales')
highest_sales
```

```python
#Using the list comprehension cause x value and y value would not match some times 

cities = [city for city, i in all_data.groupby('City Name')]

#Using matplotlib to plot the bar graph

plt.bar(cities, highest_sales['Sales'])
plt.xticks(cities, rotation='vertical')
plt.ylabel('Sales in USD($)')
plt.xlabel('US City Names')
plt.show()
```

#### Question 3 : What time should we display advertisements to maximize likelihood of customer's buying product ?

```python
all_data.head()
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
```

```python
all_data.head()
```

```python
hours = [hour for hour, i in all_data.groupby('Hour')]


plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours, rotation='vertical')
plt.title('Sales by Time')
plt.xlabel('Sales by Hours')
plt.ylabel('Number of Order)')
plt.show()
```

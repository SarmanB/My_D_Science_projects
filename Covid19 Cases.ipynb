# Web Scraping COVID-19  

```python
# Importing the dependencies 
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# Importing data from the websites
url = BeautifulSoup('https://www.worldometers.info/coronavirus/','html.parser')

soup = requests.get(url)
soup
```

```python
#Converting into .lxml format
code = BeautifulSoup(soup.text, 'lxml') 
code = code.table                          #scrapping the live market table from website
code
```

## Extracting the Table from Worldometers.com/coronavirus

```python
covid_title = code.find('tr')
covid_title
```

```python
covid_table_title = [title.text.strip() for title in covid_title]
print(covid_table_title)       # We are getting the tables as well as empty list  
```

```python
# So remove the empty table from the list we used following loop
new_covid_table_title = []
for i in covid_table_title:
    if i:
        new_covid_table_title.append(i)
    else:
        pass
new_covid_table_title
```

```python
df = pd.DataFrame(columns = new_covid_table_title)
df
```

```python
column_data = code.find_all('tr')
column_data[9]            # Checking where the value of USA is set as it is on the top of the list
```

```python
for row in column_data[9:]:        # Getting rid of the first eight cause it had other continents 
    row_data = row.find_all('td')
    individual_covid_data = [data.text.strip() for data in row_data]
    print(individual_covid_data)
    
    length = len(df)
    df.loc[length] = individual_covid_data
```
